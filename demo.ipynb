# Method: 2

```python
from pathlib import Path
from src.image_enhancer import ImageEnhancer
from src.createmask import OCTMaskGenerator

INPUT_PATH = Path("data/images/2025-05-19_12-47-45_P007_OVM_block-04_trial-06-morph_downsampled-10000.0Hz-mean.tiff")  #single image/tiff or folder path

#if single image data/masks/image.png
OUTPUT_DIR = Path("data/masks") #If you are using both method 1 and 2 dont forget to change path (etc. data/masks_2), also change OUTPUT_DIR for clean up and create csv


OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

enhancer = ImageEnhancer()                    # from image_enhancer.py
items    = enhancer.process(str(INPUT_PATH))  # list[(name, img)]
print(len(items), "frame(s) ready")
    
generator = OCTMaskGenerator(OUTPUT_DIR)
generator.create_masks(items)                # masks written + returned


```

# Clean up the masks

```python
from src.mask_cleaner import MaskCleaner   # assuming you save the class above

#Sometimes it can be better to skip this step depending on image, if small parts make easier to see trends, if so change CLEAN_MASK_DIR to OUTPUT_DIR in create mask part(cell below)

RAW_MASK_DIR  = Path(OUTPUT_DIR) # update OUTPUT_DIR = Path("mask_image_path") if you have different folder path than OUTPUT_DIR used in Method 1 or 2
CLEAN_MASK_DIR = Path(OUTPUT_DIR, "clean") # cleaned mask will be save in clean folder under OUTPUT_DIR


cleaner = MaskCleaner(min_area=10) #remove lines smaller than min_area, for method 2 use higher values 25-30 etc.

# 1) Clean every PNG in RAW_MASK_DIR and write to CLEAN_MASK_DIR
cleaner.clean_folder(RAW_MASK_DIR, CLEAN_MASK_DIR, pattern="*.png")
```

# Create CSV Data

```python
from src.line_dot_marker import LineDotMarker

marker = LineDotMarker(dot_radius=7, dot_density=0.1, color_map="random")

# Save CSV
## save_vis=True if you want to visual overlay for debugging
## If you haven't used Clean up the mask part, change CLEAN_MASK_DIR to OUTPUT_DIR
marker.process_path(CLEAN_MASK_DIR, out_dir="data/csvs", save_vis=True)

```
